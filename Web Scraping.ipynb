# Question 1

Scrape 2 pages of the reviews for a single product listing using product URL.<br /> Note: Product must have more than 1 review page. The code must be generic and should work for every product listing URL, with proper checks in place.

```python
def ReviewFinder(Reviews, Names):
    for i in range(1,6):
        products = driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[1]/div['+ str(i) +']/div[3]/div[1]')
        names = driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[1]/div['+ str(i) +']/div[2]/span[1]')
        Reviews.append(products.text)
        Names.append(names.text)

    return Reviews,Names


url = 'https://www.daraz.pk/products/waterproof-mobile-cover-i124818547-s1282674849.html?spm=a2a0e.searchlist.list.24.63144daaeIqP84&search=1'
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(url)

Reviews = []
Names = []
pgs = 2

driver.maximize_window()
time.sleep(5)

for i in range(2,pgs+2):
    Reviews,Names=ReviewFinder(Reviews, Names)
    next_button = driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[2]/div/button[2]')
    driver.execute_script("arguments[0].click();", next_button)
    time.sleep(5)

driver.close()

for i in range(len(Names)):
    Names[i]=Names[i].replace('by ','')
df = pd.DataFrame(list(zip(Names, Reviews)), columns = ['Name' , 'Comments'])
df
```
